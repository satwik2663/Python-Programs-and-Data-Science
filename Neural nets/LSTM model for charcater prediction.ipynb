```python
# Since its a character based language model, we use the input and output sequence as characters
# loading the text file
import numpy
text_file_read = open(r'Text_doc.txt','r', encoding = 'utf-8').read()
```

```python
text_file_read
```

```python
split_text = text_file_read.split('\n')
while " " in  split_text:
    split_text.remove(" ")
str_text = ' '.join(split_text)
```

```python
str_text
```

```python
# taking the charcater length as 8 and preparing the data set
data_set = []
#str_text = list(str_text)
for i in range(len(str_text)-9):
    a = str_text[i:i+10]
    data_set.append(a)
```

```python
data_set[:5]
```

```python
print('The length of the dataset is : ', len(data_set))
```

```python
# now we take the unique characters from the string
unique_characters = set(str_text)
unique_characters = list(unique_characters)
unique_characters = sorted(unique_characters)
```

```python
print(unique_characters)
print(len(unique_characters))
```

```python
# now we assign the value of unique charcters index position to uniquely identify the elements in the list
list_characters = []
for data in str_text:
    for pos, ele in enumerate(unique_characters):
        if data == ele:
            list_characters.append(pos + 1)
```

```python
final_data_set = []
for i in range(len(list_characters)-9):
    a = list_characters[i:i+11]
    final_data_set.append(a)
```

```python
final_data_set = numpy.array(final_data_set)
```

```python
final_data_set
```

```python
# splitting the final_data_set 
X = []
y = []
for data in final_data_set:
    a = []
    for i in data[:-1]:
        a.append(i)
    for j in data[-1:]:
        y.append(j)
    X.append(a)
```

```python
from keras.utils import to_categorical

sequences = [to_categorical(x, num_classes=38) for x in X]
y = to_categorical(y, num_classes=38)
```
