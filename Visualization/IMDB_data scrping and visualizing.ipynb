```python
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
```

```python
a = []
movie_names = []
movie_rating = []
movie_year = []
serial_number = []
v_name = []
v_rating = []
v_year = []
 #GETTING DATA FROM THE URL
page = requests.get('https://www.imdb.com/chart/top?ref_=nv_mv_250')
soup = BeautifulSoup(page.content, 'html.parser')

# USING SOUP OBJECT TO GET THE DATA
table_chart = soup.find('table', {'class': ['chart full-width']})
table_body = table_chart.find('tbody',{'class':'lister-list'})

#FINDING THE TABLE ELEMENTS
tr = table_body.find_all('tr')
for i in tr:
    x = i.find('td',{'class':'titleColumn'})
    y = x.find('a')
    movie_names.append(y.text)
    year = i.find('span',{'class':'secondaryInfo'})
    movie_year.append(year.text[1:5])
   

for i in tr:
    x = i.find('td',{'class':'ratingColumn imdbRating'})
    movie_rating.append(x.text[1:4])


movie = zip(movie_names, movie_rating, movie_year)

# CONVERTING THE LIST OF MOVIE NAME, MOVIE YEAR AND RATING TO A DATA FRAME
df = pd.DataFrame({'Movie names': movie_names,'Movie rating': movie_rating,'Movie year': movie_year})
df.set_index('Movie names', inplace = True) 
print(df)


list_sorted_by_year = sorted(movie, key=lambda tup: tup[2])

# SORTING MOVIES BY YEAR
for name, rating, year in list_sorted_by_year:
    v_name.append(name)
    v_rating.append(float(rating))
    v_year.append(int(year))

#PLOTTING THE TABLE ELEMENTS
plt.rcParams['figure.figsize'] = [25,10]
plt.bar(v_year, v_rating,color = 'r', align='center', width=0.4) 
plt.xlabel('Year', fontsize=18)
plt.ylabel('Movie Rating', fontsize=16)
plt.xticks(v_year,rotation=90)
#plt.xticks()
plt.grid()
```
