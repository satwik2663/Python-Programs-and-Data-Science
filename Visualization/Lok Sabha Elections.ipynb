# LOK SABHA 2019

```python
links_2019 = []
serial_number_2019 = []
candidates_2019 = []
party_2019 = []
criminal_cases_2019 = []
education_2019 = []
age_2019 = []
total_asset_2019 = []
liabilities_2019 = []
state_2019 = []
list_rows_2019 = []
year_2019 = []
link_can_list_2019 = []
```

```python

options = Options()
ua = UserAgent()
userAgent = ua.random

options.add_argument(f'user-agent={userAgent}')
driver = webdriver.Chrome(chrome_options=options, executable_path = '.\chromedriver.exe')

driver.get('http://www.myneta.info/LokSabha2019/')
html = driver.page_source
soup = BeautifulSoup(html)
name_containers = soup.find_all('h5', class_ = 'title')


for i in name_containers:
    links_2019.append("https://www.myneta.info/LokSabha2019/" + i.find('a')['href'])        
    
for i in links_2019[1:2]:
    val_rows = []
    driver.get(i)
    html = driver.page_source
    soup = BeautifulSoup(html)
    state_name = soup.find('div',class_ ='items')
    table = soup.findChildren('table')[2]
    rows = table.find_all('tr')
    for row in rows:
        cols=row.find_all('td')
        cols=[x.text.strip() for x in cols]
        link_can = row.find_all('a',href = True)
        link_can_list_2019.append('http://www.myneta.info/LokSabha2019' + str(link_can))
        cols.append(state_name.text)
        cols.append('2019')
        list_rows_2019.append(cols)
```

```python
for i, j in enumerate(link_can_list_2019):
    if j == 'http://www.myneta.info/LokSabha2019[]':
        link_can_list_2019.pop(i) 
    else:
        continue 
```

```python
for list_ele in list_rows_2019[1:]:
    if len(list_ele) == 10:
        #serial_number_2019.append(list_ele[0])
        candidates_2019.append(list_ele[1])
        party_2019.append(list_ele[2])
        criminal_cases_2019.append(list_ele[3])
        education_2019.append(list_ele[4])
        age_2019.append(list_ele[5])
        total_asset_2019.append(list_ele[6])
        liabilities_2019.append(list_ele[7])
        state_2019.append(list_ele[8])
        year_2019.append(list_ele[9])
    else:
        continue
```

```python
dataset_candidates_2019 = pd.DataFrame()
#dataset_candidates_2019['Sno'] = serial_number
dataset_candidates_2019['candidates'] = candidates_2019
dataset_candidates_2019['party'] = party_2019
dataset_candidates_2019['criminal_cases'] = criminal_cases_2019
dataset_candidates_2019['education'] = education_2019
dataset_candidates_2019['age'] = age_2019
dataset_candidates_2019['total_asset'] = total_asset_2019
dataset_candidates_2019['liabilities'] = liabilities_2019
dataset_candidates_2019['state'] = state_2019
dataset_candidates_2019['state'] = state_2019
dataset_candidates_2019['year'] = year_2019
dataset_candidates_2019['links'] = link_can_list_2019
```

## Cleaning the dataset

```python
copy = result_df.copy()
```

```python
# removing values from state
result_df['state'] =  [x.replace('Home→Lok Sabha 2019→','') for x in result_df['state']]
result_df['state'] =  [x.replace('Home→Loksabha 2014→','') for x in result_df['state']]
result_df['state'] =  [x.replace('Home→Lok Sabha 2009→','') for x in result_df['state']]
```

```python
# removing irrelevent values from total asset and liabilities column
result_df['total_asset'] =  [x.replace(',','') for x in result_df['total_asset']]
result_df['liabilities'] =  [x.replace(',','') for x in result_df['liabilities']]
```

```python
result_df['total_asset'] = result_df['total_asset'].str.split('~').str[0]
result_df['liabilities'] = result_df['liabilities'].str.split('~').str[0]
```

```python
#state column
result_df['state'].unique()
```

```python
# age column
result_df['age'].unique()
result_df.loc[result_df['age'] == 'Unknown','age'] = '0'
result_df['age'].astype(int)
```

```python
# 'age' has some missing values. So, we can replace it with median for each year
```

```python
result_df.loc[(result_df['age'] == 'Unknown') & (result_df['year'] == '2014'),'age'] = result_df.loc[(result_df['age'] != 'Unknown') & (result_df['year'] == '2014'),'age'].median()
result_df.loc[(result_df['age'] == 'Unknown') & (result_df['year'] == '2010'),'age'] = result_df.loc[(result_df['age'] != 'Unknown') & (result_df['year'] == '2010'),'age'].median()
result_df.loc[(result_df['age'] == 'Unknown') & (result_df['year'] == '2019'),'age'] = result_df.loc[(result_df['age'] != 'Unknown') & (result_df['year'] == '2019'),'age'].median()
```
